```python

from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris = load_iris()
```

```python
X = iris.data
```

```python
X
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
inertias = []
k_cluster = range(1, 12)
for k in k_cluster:
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(X)
  inertias.append(kmeans.inertia_)

plt.plot(k_cluster, inertias, marker='o')
plt.xlabel('Nombre de Clusters')
plt.ylabel('Inertia')
plt.title('ELbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=20)
kmeans.fit(X)
```

```python
y_pred = kmeans.predict(X)
center_values = kmeans.cluster_centers_[y_pred]
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
rmse = mean_squared_error(X, center_values, squared=False)
mae = mean_absolute_error(X, center_values)

print("RMSE:", rmse)
print("MAE:", mae)
```

```python
plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.scatter(center_values[:,0], center_values[:,1], marker='x', color='blue', s=200, label='Centre de clusters')
plt.xlabel('SapalWidth')
plt.ylabel('PetalWidth')
plt.title('Resultat du clustering avec KMeans')
plt.legend()
plt.show()
```

** Methode de l'agglomerative Clustering **

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
Y = iris.data
```

```python
Y
```

```python
from scipy.spatial.distance import pdist, squareform
distance = squareform(pdist(Y))
```

```python
distance
```

```python
from sklearn.cluster import AgglomerativeClustering
agglo = AgglomerativeClustering(n_clusters=3,linkage='ward')
agglo.fit_predict(distance)
```

```python
x_pred = agglo.fit_predict(distance)
```

```python
from sklearn.metrics import silhouette_score
labels = agglo.labels_
s_score = silhouette_score(distance, labels)
print("Sihouette score :", s_score)
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
plt.scatter(Y[:, 0], Y[:, 1], c=labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clustering Result')
plt.show()
```
